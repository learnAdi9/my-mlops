```python
# Preparing data function - Pipe2!

def prep_data(data_source):
  df = pd.read_parquet(data_source)
  # Creating duration variable & examining it !
  df['duration'] = df['tpep_dropoff_datetime'] - df['tpep_pickup_datetime']
  df['duration'] = df['duration'].apply(lambda dur: round(dur.total_seconds() / 60 , 2))
  df = df[(df.duration >= 1) & (df.duration <= 60)]
  # cat-features from EDA
  cat_features = ['PULocationID', 'DOLocationID']
  df[cat_features] = df[cat_features].astype(str)

  return df
```

```python
# Preparing Data using Mar-2023 data (Yellow Taxi)!
df = prep_data(data_source = 'https://d37ci6vzurychx.cloudfront.net/trip-data/yellow_tripdata_2023-03.parquet')
len(df)
```

```python
# Use prep_data to return DictVect & trained Model - Pipe3!

def vect_and_train(df):
  # Dictionary Vectorizer & X_train data
  dVec = DictVectorizer()
  feature_dicts = df[['PULocationID', 'DOLocationID']].to_dict(orient = 'records')
  X_train = dVec.fit_transform(feature_dicts)
  # Obtaining target and creating target vector !
  y_train = df['duration'].values.copy()

  # Training a Simple Linear Reg Model !
  lr = LinearRegression()
  model = lr.fit(X_train, y_train)
  print("Intercept: ",model.intercept_)

  return dVec, model
```

```python
# Call and Check
_ = vect_and_train(df)
```

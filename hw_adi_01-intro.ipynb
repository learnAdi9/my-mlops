## EDA

```python
# Checking for features with NULLs !
data_df.isnull().sum()
```

```python
# Check in categoricals!!
data_df.store_and_fwd_flag.nunique(), data_df.payment_type.nunique(),data_df.VendorID.nunique()
data_df.RatecodeID.nunique(),data_df.PULocationID.nunique(),data_df.DOLocationID.nunique()
```

```python
# Examining Categories!!
data_df.store_and_fwd_flag.value_counts(), data_df.payment_type.value_counts(), data_df.VendorID.value_counts(), data_df.RatecodeID.value_counts()
```

```python
# Examining some Numerical features with NULLs !
data_df[data_df.passenger_count.isnull()].sample(10)
```

```python
# EDA for duration variable !
len(data_df[data_df['tpep_dropoff_datetime'] < data_df['tpep_pickup_datetime']])

# Creating duration variable & examining it !
data_df['duration'] = data_df['tpep_dropoff_datetime'] - data_df['tpep_pickup_datetime']
data_df['duration'] = data_df['duration'].apply(lambda dur: round(dur.total_seconds() / 60 , 2))

round(data_df['duration'].describe(),2) # Q2 - Ans
```

```python
# Outliers in duration variable !
data_df = data_df[(data_df.duration >= 1) & (data_df.duration <= 60)]
round(len(data_df)/3066766,2)*100 # Q3 - Ans
```

```python
# Business selected features for the Base Model
cat_features = ['PULocationID', 'DOLocationID']
data_df[cat_features] = data_df[cat_features].astype(str)

# Dictionary Vectorizer & X_train data
dVec = DictVectorizer()

feature_dicts = data_df[cat_features].to_dict(orient = 'records')
X_train = dVec.fit_transform(feature_dicts)
X_train # Q5 - Ans
```

```python
# Obtaining target and creating target vector !
y_train = data_df['duration'].values.copy()
len(y_train)
```

```python
# Training a Simple Linear Reg Model !

lr = LinearRegression()
jan_model = lr.fit(X_train, y_train)
y_pred = jan_model.predict(X_train)

# Training RMSE: Q5 - Ans !
mean_squared_error(y_train, y_pred, squared = False)
```

```python
# Data Process & Prep Function !
def prep_data(data_source):
  df = pd.read_parquet(data_source)
  # Creating duration variable & examining it !
  df['duration'] = df['tpep_dropoff_datetime'] - df['tpep_pickup_datetime']
  df['duration'] = df['duration'].apply(lambda dur: round(dur.total_seconds() / 60 , 2))
  df = df[(df.duration >= 1) & (df.duration <= 60)]
  # cat-features from EDA
  cat_features = ['PULocationID', 'DOLocationID']
  df[cat_features] = df[cat_features].astype(str)

  return df
```

```python
# Preparing validation Data using Feb-2023 data
df_val = prep_data(data_source = 'https://d37ci6vzurychx.cloudfront.net/trip-data/yellow_tripdata_2023-02.parquet')
```

```python
# Business selected features for model validation !
cat_features = ['PULocationID', 'DOLocationID']

feature_dicts = df_val[cat_features].to_dict(orient = 'records')
X_val = dVec.transform(feature_dicts)

# Obtaining target and creating target vector for Validation !
y_val = df_val['duration'].values.copy()
```

```python
# Use the Trained model of Jan to Predict for validation Data !
y_pred = jan_model.predict(X_val)

# Validation RMSE: Q6 - Ans !
mean_squared_error(y_val, y_pred, squared = False)
```
